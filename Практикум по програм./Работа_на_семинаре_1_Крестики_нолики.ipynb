```python
pole = [1,2,3,4,5,6,7,8,9]
win_pole = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[0,4,8],[2,4,6]]
def print_pole():
    print(pole[0], end = ' ')
    print(pole[1], end = ' ')
    print(pole[2])
    print(pole[3], end = ' ')
    print(pole[4], end = ' ')
    print(pole[5])
    print(pole[6], end = ' ')
    print(pole[7], end = ' ')
    print(pole[8])

#Функция, чтобы сделать ход
def hod(hod,symbol):
    ind = pole.index(hod)
    pole[ind]=symbol

#Функция для получения результата игры
def result():
    win = ''
    for i in win_pole:
      if pole[i[0]] == "X" and pole[i[1]] == "X" and pole[i[2]] == "X":
            win = "X"
      elif pole[i[0]] == "O" and pole[i[1]] == "O" and pole[i[2]] == "O":
            win = "O"
    return win

#Поиск линии с победным числом Х и О для компьютера
def check_line(S_O, S_X):
  step = ''
  for i in win_pole:
    x = 0
    o = 0
    for j in range(0,3):
      if pole[i[j]] == 'O':
        o+=1
      if pole[i[j]] == 'X':
        x+=1
    if o == S_O and x == S_X:
      for j in range(0,3):
        if pole[i[j]] != 'O' and pole[i[j]] != 'X':
          step = pole[i[j]]
  return step

#Компьютер выбирает наиболее выигрышный ход
def choice():
  step = ''
  #Если на линии стоит 2 своих фигуры, а чужих нет
  step = check_line(2,0)
  #Если на линии стоит 1 своя фигура, а чужих нет
  if step == '':
    step = check_line(1,0)
  #Если центр пустой, то ставим свою фигуру
  if step == '':
    if pole[4] != 'X' and pole[4] != 'O':
      step = 5
  #В противном случае ставим фигуру на 1 клетку поля
  if step == '':
    if pole[0] != 'X' and pole[0] != 'O':
      step = 1

  return step

#Основное тело игры
print('Раунд 1')
game = False
h = 1
while game == False:
  print_pole()
  if h==1:
      symbol = 'X'
      step = int(input('Введите координату для крестика: '))
  else:
      print('Компьютер делает ход: ')
      symbol = 'O'
      step = choice()
  #Если компьютер нашел, куда сделать ход
  if step != '':
    hod(step, symbol)
    win1 = result()
    if win1 != '':
      game = True
    else:
      game = False
  else:
    print('Ничья')
    game = True
    win1 = 'ничья'
  h = not(h)
print('Выиграл: ', win1)
#Завела счетчики, чтоб считать победы 
x=o=n=0
if win1 == 'X':
  x += 1
elif win1 == 'O':
  o += 1
else:
  n +=1

#В каждом раунде перерисовываем поле, чтобы игра началась с самого начала
print('Раунд 2')
pole = [1,2,3,4,5,6,7,8,9]
win_pole = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[0,4,8],[2,4,6]]
game = False
h = 1
while game == False:
  print_pole()
  if h==1:
      symbol = 'X'
      step = int(input('Введите координату для крестика: '))
  else:
      print('Компьютер делает ход: ')
      symbol = 'O'
      step = choice()
  #Если компьютер нашел, куда сделать ход
  if step != '':
    hod(step, symbol)
    win2 = result()
    if win2 != '':
      game = True
    else:
      game = False
  else:
    print('Ничья')
    game = True
    win2 = 'ничья'
  h = not(h)
print('Выиграл: ', win2)
if win2 == 'X':
  x += 1
elif win2 == 'O':
  o += 1
else:
  n +=1

print('Раунд 3')
pole = [1,2,3,4,5,6,7,8,9]
win_pole = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[0,4,8],[2,4,6]]
game = False
h = 1
while game == False:
  print_pole()
  if h==1:
      symbol = 'X'
      step = int(input('Введите координату для крестика: '))
  else:
      print('Компьютер делает ход: ')
      symbol = 'O'
      step = choice()
  #Если компьютер нашел, куда сделать ход
  if step != '':
    hod(step, symbol)
    win3 = result()
    if win3 != '':
      game = True
    else:
      game = False
  else:
    print('Ничья')
    game = True
    win3 = 'ничья'
  h = not(h)
print('Выиграл: ', win3)
if win3 == 'X':
  x += 1
elif win3 == 'O':
  o += 1
else:
  n +=1

#Теперь по счетчикам отпределяю, кто победил
if x > o and x > n:
  print('По результатам трёх раундов выиграл КРЕСТИК')
elif o > x and o > n:
  print('По результатам трёх раундов выиграл НОЛИК')
else:
  print('По результатам трёх раундов победила ДРУЖБА')

#КОНЕЦ! Всё работает отлично! Крестик также может победить!





```
