1. Считайте слова из файла `litw-win.txt` и запишите их в список `words`. В заданном предложении исправьте все опечатки, заменив слова с опечатками на ближайшие (в смысле расстояния Левенштейна) к ним слова из списка `words`. Считайте, что в слове есть опечатка, если данное слово не содержится в списке `words`. 

```python
words = dict()
f = open('/content/litw-win.txt', encoding = 'cp1251')
for i in f.readlines():
  word = i.split()[1]
  cnt = i.split()[0]
  words[word] = int(cnt)
words
```

```python
import nltk
text = '''с велечайшим усилием выбравшись из потока убегающих людей Кутузов со свитой уменьшевшейся вдвое поехал на звуки выстрелов русских орудий'''
new_text = ''
change = []
for word in nltk.word_tokenize(text):
  if word not in words.keys():
    compare = []
    for i in words.keys():
      price = nltk.edit_distance(word, i)
      if len(compare) == 0:
        compare = [i, price]
        continue
      if price < compare[1]:
        compare = [i, price]
    change += [(word, compare[0])]
    word = compare[0]
  new_text += word + ' '
print(new_text)
print(change)
```

1.3 Напишите функцию, которая для заданного слова `word` возвращает `k` ближайших к нему слов из списка `words` (близость слов измеряется с помощью расстояния Левенштейна)

```python
def wordd(word, k, words):
  words_tuple = list()
  for i in words:
    words_tuple.append(tuple([edit_distance(word, i), i]))
  words_tuple.sort(key = lambda x: x[0])
  result = list()
  for i in words_tuple[:k]:
    result.append(i[1])
  return result
wordd("delicious", 4, words)
```

3.1 Выберите случайным образом 5 рецептов из набора данных. Представьте описание каждого рецепта в виде числового вектора при помощи `TfidfVectorizer`

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
f = pd.read_csv('/content/preprocessed_descriptions.csv', sep = ',')
random_recipes = f.sample(5)
random_recipes["preprocessed_descriptions"] = random_recipes["preprocessed_descriptions"].map(sent_tokenize)
tv = TfidfVectorizer()
text = list()
for index, row in random_recipes.iterrows():
  text.append(row["preprocessed_descriptions"][0])
tv2 = tv.fit_transform(text)
tv3 = tv2.toarray()
tv3
```

3.2 Вычислите близость между каждой парой рецептов, выбранных в задании 3.1, используя косинусное расстояние (`scipy.spatial.distance.cosine`) Результаты оформите в виде таблицы `pd.DataFrame`. В качестве названий строк и столбцов используйте названия рецептов.

```python
tab = pd.DataFrame(index=random_recipes['name'], columns = random_recipes['name'])
for i in range(len(random_recipes)):
  for j in range(i, len(random_recipes)):
    tab.iloc[i][j] = 1 - distance.cosine(tv3[i], tv3[j])
    tab.iloc[j][i] = 1 - distance.cosine(tv3[i], tv3[j])
tab
```
